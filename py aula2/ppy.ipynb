# Carregando Ferramentas

```python
import pandas as pd
# Assuming the data is in a CSV file named 'stock_data.csv'
df = pd.read_csv('BTC-USD(1).csv', encoding='latin1')
print(df.head())
import matplotlib.pyplot as plt
```

#

# 1. Qual o preço médio de fechamento para todo o período?

```python
average_closing_price = df['Close'].mean()
print("Média dos preços de fechamento:", average_closing_price)

# Create a horizontal line to represent the average closing price
plt.axhline(y=average_closing_price, color='r', linestyle='-', label='Média dos preços de fechamento')

# Plot the closing price over time
plt.plot(df['Date'], df['Close'], label='Preço de fechamento')
plt.xlabel('Date')
plt.ylabel('Preço de fechamento')
plt.title('Evolução do preço ao longo do tempo')
plt.legend()
plt.show()
```

O gráfico mostra a variação do preço de fechamento de um ativo ao longo do tempo, comparando-o com a média de todos os preços. A linha vermelha representa a média, e a linha azul representa o preço de fechamento diário.

#

# 2. Qual o dia com o maior valor de negociação?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que 'df' é seu DataFrame e as datas estão no formato "%Y-%m-%d"
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.set_index('Date', inplace=True)

# Encontrar o dia com maior volume e o valor correspondente
max_volume_day_index = df['Volume'].idxmax()
max_volume = df['Volume'].max()

# Converter o índice para uma data legível
max_volume_day = max_volume_day_index.strftime('%Y-%m-%d')

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Volume'], label='Volume diário')
plt.axhline(y=max_volume, color='red', linestyle='--', label='Maior volume')
plt.xlabel('Data')
plt.ylabel('Volume')
plt.title('Volume de Negociação ao Longo do Tempo')
plt.legend()
plt.grid(True)
plt.show()

print("Dia com o maior volume:", max_volume_day)
print("Valor do maior volume:", max_volume)
```

O gráfico mostra a quantidade de negociações de um ativo ao longo do tempo. O pico mais alto indica o dia em que houve o maior volume de negociações, que foi em 26 de fevereiro de 2021.

#

# 3. Quais são os retornos diários?

```python
# Calculate the daily returns
df['Daily_Return'] = df['Close'].pct_change()

# Criar um histograma dos retornos diários
plt.hist(df['Daily_Return'], bins=20, density=True)
plt.xlabel('Retorno Diário')
plt.ylabel('Densidade')
plt.title('Distribuição dos Retornos Diários')
plt.show()
```

O gráfico mostra a distribuição dos retornos diários de um investimento. Cada barra representa a frequência com que o investimento teve um determinado retorno em um dia. A forma da distribuição indica a volatilidade do investimento.

#

# 4. Qual o preço de fechamento ao longo do tempo?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar dados da planilha Excel
df = pd.read_csv('BTC-USD(1).csv', encoding='latin1')

# Mostrar as primeiras linhas para verificar o carregamento correto


# Converter a coluna 'Date' para o tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Configurar o tamanho da figura
plt.figure(figsize=(12, 6))

# Plotar o preço de fechamento ao longo do tempo
plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', color='b', label='Preço de Fechamento')

# Adicionar rótulos aos eixos e título
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Preço de Fechamento ao Longo do Tempo')

# Adicionar uma grade para facilitar a leitura
plt.grid(True)

# Exibir a legenda
plt.legend()

# Mostrar o gráfico
plt.show()
```

O gráfico mostra a variação do preço de fechamento do Bitcoin ao longo dos anos. Cada ponto representa o preço de fechamento em um determinado dia, e a linha conecta esses pontos para mostrar a tendência geral do preço."
